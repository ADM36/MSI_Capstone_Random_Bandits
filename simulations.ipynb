# Experiments for the final presentation/paper

```python
labeler_sets = {
    'all_0.5' : [0.5, 0.5, 0.5, 0.5, 0.5, 0.5],
    'all_0.6' : [0.6, 0.6, 0.6, 0.6, 0.6, 0.6],
    'all_0.7' : [0.7, 0.7, 0.7, 0.7, 0.7, 0.7],
    'all_0.8' : [0.8, 0.8, 0.8, 0.8, 0.8, 0.8],
    'all_0.9' : [0.9, 0.9, 0.9, 0.9, 0.9, 0.9],
    'one_0.9' : [0.5, 0.5, 0.5, 0.5, 0.5, 0.9],
    'two_0.9' : [0.5, 0.5, 0.5, 0.5, 0.9, 0.9],
    'three_0.9' : [0.5, 0.5, 0.5, 0.9, 0.9, 0.9],
    'four_0.9' : [0.5, 0.5, 0.9, 0.9, 0.9, 0.9],
    'five_0.9' : [0.5, 0.9, 0.9, 0.9, 0.9, 0.9],
}

acc = []
percent_conf = []
ncs = {}
nis = {}

for name,labeler_set in labeler_sets.items():
    test = pipeline.SimulationPipeline(labelers=labeler_set,levers={'a':0.2,'b':0.3,'c':0.4,'d':0.5})
    test.run_simulation(500)
    test.plot_bandits(savefile=f'final_images/paper_{name}.png')
    acc.append(np.array(labeler_set).mean())
    ncs[name] = (test.total_reward[-1])
    nis[name] = (test.total_labeled[-1])
    percent_conf.append(test.total_reward[-1]/test.total_labeled[-1])
```

```python
conf = percent_conf.copy()
acc2 = acc.copy()
```

```python
plt.scatter(conf,acc2)
```

### Interpretation:
So, we can create this curve which approximates how "accurate" the labelers are finding correct positive samples, as determined by the percent conf.

```python
# a,b = np.polyfit(np.log(percent_conf), acc, 1)

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def func(x, a, b):
    return a * np.log(x) + b

popt, pcov = curve_fit(func, conf, acc2)
```

```python
a,b = popt
conf.sort()
conf
```

```python
fig = plt.figure(figsize=[9,9])
ax = fig.add_subplot(111)

x = np.arange(0.05,1.9,0.001)
ax.scatter(percent_conf,acc,s=150, color='red',label="Simulated Experiments")
ax.plot(x, a*np.log(x)+b, lw=3, color='black', label="LLS Logarithmic Fit")

ax.set_ylim(0.4,1)
ax.set_xlim(0,1)
ax.set_xlabel("Confidence Ratio",size=18)
ax.set_ylabel("Average Accuracy of Synthetic Labelers",size=18)
ax.tick_params(axis='both', labelsize=13)
ax.set_title("Accuracy of Experimental Labelers vs. Confidence Ratio",size=22)
plt.legend(loc=0, prop={'size': 16})

ax.text(0.45, 0.55, r"y = 0.201 log(x) + 1.01", size=12, transform=ax.transAxes)

plt.savefig("./final_images/paper_acc_v_conf_ratio.png",dpi=150)
```
